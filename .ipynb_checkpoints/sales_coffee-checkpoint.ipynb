```python
df = pd.read_csv('../sales coffee/index.csv')
df
```

```python
df.isnull()
```

```python
df_nuevo=df.dropna()
```

```python
df_nuevo
```

```python
#Se realiza un pivot a la tabla para que se indexe por el producto y la fecha de venta
df_pivot=df_nuevo.pivot_table(index='coffee_name',columns='date', values='money')
df_pivot
```

```python
#Se agrupa por la fecha en que se vendió el producto y el intervalo de tiempo, identificando los valores que luego se suman
df_segmentado=df_nuevo.groupby(['date','coffee_name']).agg(
    {'datetime':['min','max'],
     'money':lambda x:[i for i in x]},
    
)                                                        
df_segmentado
```

```python
df_segmentado=df_nuevo.groupby(['date','coffee_name']).agg(
    {'datetime':['min','max'],
     'money':'sum','coffee_name':'count'})
df_segmentado
```

```python
#la consulta anterior se aplana para saber el total del producto vendido en cada fecha
df_money = df_segmentado['money']['sum'].unstack()
df_money 
```

```python
# Graficamos los datos conocer la tendencia del producto vendida en el tiempo indicado
fig, ax = plt.subplots(figsize=(10, 6))
df_money.plot(ax=ax)

ax.set_title('Suma de dinero por tipo de café a lo largo del tiempo')
ax.set_ylabel('Suma de dinero')
ax.set_xlabel('Fecha')

plt.show()
```
